### Advanced Filtering in pandas

```python
import pandas as pd 

df = pd.read_csv('../../../datasets/student.csv')
df
```

```python
# Multiple conditions

df[(df['Semester'] == 'IX') & (df['gender'] == 'm')]
```

```python
# Use of isin operators

df[df['place'].isin(['lalitpur'])]
```

```python
# Use of .str.contains()

df[df['Name'].str.contains('ita',case=True)]
```

```python
# Use of between operators

df[df['Marks'].between(70,99)]
```

```python
# Use of .query statement

df.query("grade == 'A' and Marks > 90")
```

```python
# Sorting by Multiple Columns

df.sort_values(by=['grade', 'Marks'], ascending=[True, False])
```
