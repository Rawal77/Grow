### Filling null values

```python
import pandas as pd
import numpy as np 

df = pd.read_csv("../../../datasets/student.csv")
df
```

### Dropping the missing values

```python
df
```

```python
droped_value = df.dropna()
droped_value
```

```python
# Dropping a specific column which have NaN

droped_column = df.dropna(subset=['Semester','place'])
droped_column
```

```python
df['Semester'] = df['Semester'].ffill()
df
```

```python
df['place'] = df['place'].bfill()
df
```

```python
df['gender'] = df['gender'].fillna('Unknown')
df
```
