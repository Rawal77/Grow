### Handling Null Values

```python
import numpy as np
import pandas as pd

data = {
    "Name" : ["Ram","Shyam","Gita"],
    "Age" : [12,13,11],
    "Gender" : ['m',np.nan,'f']
}

df = pd.DataFrame(data)
df
```

```python
df = pd.read_csv('../../../datasets/student.csv')
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df['Semester'].isnull()]
```

```python
df.notnull().sum()
```

```python
df[df['Semester'].notnull()]
```

```python
df[df[['Semester','Name']].notnull().all(axis=1)]
```
