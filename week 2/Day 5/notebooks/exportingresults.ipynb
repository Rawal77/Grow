### Exporting results to csv and excel files and excel sheets

```python
import pandas as pd 

data = {
    "Name" : ['Ram','Shyam','Hari','Ghanshyam','Sita','Gita','Rita'],
    "Electives" : ['Cyber Security','IS','Cyber Security','ML','ML','IS','IS'],
    "Gender" : ['M','M','M','M','F','F','F'],
    "Marks Scored" : [70,80,50,32,21,65,75]
}

df = pd.DataFrame(data)
df
```

```python
grouped = df.groupby('Electives')['Marks Scored'].agg(['min','mean','max']).reset_index()
grouped
```

```python
grouped_gender = df.groupby(['Electives','Gender'])['Marks Scored'].agg(['min','mean','max']).reset_index()
grouped_gender
```

```python
# Exporting a csv file

grouped.to_csv('../reports/avg_marks_by_dept.csv',index=False)
```

```python
grouped_gender.to_excel('../reports/avg_marks_by_dept_gender.xlsx',index=False)
```
