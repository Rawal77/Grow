###  Prepare Sample DataFrames

```python
import pandas as pd
```

```python
left = pd.DataFrame({
    'Name' : ['Ram','Rama','Gita','Hari','Sita'],
    'ID' : [1,2,3,4,5,]
})

right = pd.DataFrame({
    'ID' : [3,5,6,7,8],
    'Subject' : ['Math','Computer','Science','English','Nepali']
})
```

```python
left
```

```python
right
```

```python
# Default (inner) join

inner = pd.merge(left,right,on='ID',how='inner')
inner
```

```python
# Left join

left_join = pd.merge(left,right,on='ID',how='left')
left_join
```

```python
# Right join 

right_join = pd.merge(left,right,on='ID',how='right')
right_join
```

```python
# Outer join


outer_join = pd.merge(left,right,on='ID',how='outer')
outer_join
```

```python
# cross join 

cross = left.merge(right,how='cross')
cross
```
