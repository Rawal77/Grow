### Bar plot

A bar plot (bar chart) displays categorical data with rectangular bars where:

    - Bars can be vertical (plt.bar()) or horizontal (plt.barh())

Why is a barplot needed?

    - Compare values accross different categories
    - Visualize frequency or counts of categorical data.
    - display discrete data (not continuous)

Best use cases for Bar plot

    - Categories are limited (not too many)
    - You want to highlight difference between categories
    - You need to compare individual data points, not distributions

When to Avoid Bar Plots?

    - Data is continuous (use line plot or histogram needed)
    - You have too many categories
    - You need to show distribution or variability (use box plot or violin plot)
    - Time series with frequent intervals (use line chart instead)

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x = ['Python','C','C++','Java','Javascript']
y = [4,3,2,4,5]

plt.xlabel('Programming language')
plt.ylabel('Ratings')
plt.title('Rating of programming languages')
plt.bar(x,y)
```

```python
plt.bar(x,y,width=0.5,color=['r','g','b','y','g'])
```

```python
plt.bar(x,y,width=0.5,color=['r','g','b','y','g'],align='edge')
```

```python
plt.bar(x,y,edgecolor='r')
```

```python
plt.bar(x,y,color='pink',edgecolor='g',linewidth=2)
```

```python
plt.bar(x,y,color='orange',edgecolor='g',linewidth=2,linestyle=':')
```

```python
plt.bar(x,y,edgecolor='green',color='red',linewidth=2,linestyle='--',alpha=0.7)
```

```python
days = [1, 2, 3, 4, 5]
temperature = [30.1, 32.5, 29.8, 31.0, 30.7]

plt.bar(days, temperature)
plt.title("Temperature Over 5 Days")
plt.xlabel("Day")
plt.ylabel("Temperature (C)")
plt.show()

# Youre better off seeing trends, not category comparisons.
```

```python
x = ['Apple','Banana','Litchi','Mango','Orange']
y = [4,5,6,3,4]

plt.bar(x,y,color=['r','g'],alpha=0.8,label='Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('No of fruits sold')
plt.title('Number of fruits sold Representation')
plt.legend()
```
