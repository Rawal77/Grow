### Box and Whisker plot

What is a Box and whisker plot?

    A Box plot displays the distribution of a dataset by showing:
    Min, Max, 1st quartile, 2nd quartile(Median), 3rd quartile, Maximum, Outliers

    It helps in understanding spread, skewness, outliers in your data

When to use a Box Plot?

    Use When

    You want to compare distributions between groups
    You want to see outliers
    You want to summarize variations using just 5 key values

Avoid When???

    Dataset is too small -- may not represent spread well
    You need to see exact values (use histogram or scatter instead)

```python
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
```

```python
x = [10,20,30,40,50,60,70,110]
plt.grid(True)
plt.boxplot(x)
```

```python
data = np.random.randn(100)  # Normal distribution

plt.boxplot(data)
plt.title("Basic Box Plot")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
group1 = np.random.normal(0,1,100)
group2 = np.random.normal(1,1.5,100)
group3 = np.random.normal(2,0.5,100)

plt.boxplot([group1,group2,group3],labels=['A','B','C'])
plt.title('Grouped Box plot')
plt.show()

```

```python
plt.boxplot([group1,group2,group3],labels=['A','B','C'],vert=False)
```
