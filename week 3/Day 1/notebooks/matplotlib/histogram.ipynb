### Histogram

What is histogram?
    
    A histogram is a plot used to visualize the distribution of a continuous variable by dividing the range into interval called bins, and then counting how many values fall into each bin.

    It is used to understand the shape of data-- eg normal distribution, skewness, spread

When to use a histogram?

    Best Cases
    
        - To see frequency distibution
        - To analyze skewnees or spread of numeric data
    
    Avoid When:

        - Data is categorical -> use bar plot
        - You are comparing small discrete values -> scatter or bar might be better
        - You need exact value counts (use bar plot or value_counts)

```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data = [1, 2, 2, 3, 4, 5, 6, 7, 8, 9]

plt.hist(data, bins=5, edgecolor='black')
plt.title("Histogram with 5 bins")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

```python
data = [1, 2, 2, 3, 4, 5, 6, 7, 8, 9]
bins_value = [1, 3, 5, 7, 9]
plt.xticks(bins_value)
plt.hist(data,bins=bins_value,edgecolor='black')
```

```python
ages = [13, 15, 18, 22, 25, 29, 34, 36, 45, 52, 60]
bins = [0, 18, 30, 45, 60]

plt.hist(ages, bins=bins, edgecolor='black')
plt.xticks(bins)
plt.title("Age Groups")
```

```python
plt.hist(ages,bins=bins,histtype="step")
```
