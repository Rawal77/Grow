### Line Plot

What is a line plot?

            A line plot is used to show trends over continuous data, especially time series or ordered values.
            It connects data points with lines, which helps visualize:
             . patterns
             . fluctuations
             . comparison over time or order

When to use Line Plot

    Best Scenarios
        . Data with order (Time,Sequence)
        . Observing trends or progress
        . Comparing multiple data series
    
    Avoid Line Plots
        . Data is categorical or unordered
        . You have too many data points
        . You need to show discrete values only (bar/scatter)

```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
x = [1,2,3,4,5]
y= [10,20,15,25,30]

plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
```

```python
plt.plot(x,y,c='g',linestyle='--',linewidth=2,marker='o')
plt.grid(True)
```

```python
x = [1, 2, 3, 4, 5]

y1 = [10,12,14,18,29]
y2 = [12,10,18,30,25]

plt.plot(x,y1,c='r',label='Line A',marker='*')
plt.plot(x,y2,c='b',label='Line B',marker='*')
plt.plot(marker='*')
plt.grid(True)

plt.title('Multiple line plot')
```
