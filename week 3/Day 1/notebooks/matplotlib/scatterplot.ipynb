### Scatterplot

What is a scatter plot?

    A scatter plot shows individual data point on a 2D graph using dots.

Why is a scatter plot needed?

    You want to see if there is a relationship between two variables
    You want to detect patterns, clusters or outliers
    You are visualizing continuous variables (eg height vs weight, age vs salary)

Best use cases for scatter plots

    Visualizing correlation (positive/negative/none)
    Showing data density or clusters
    Identifying anamolies or outliers

When to avoid scatter plots

    You're comparing categories (use bar plot instead)
    You only have one variable (use histogram or line plot)
    Too much overlaping (hard to read)

```python
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
```

```python
arr1 = np.random.randint(1,10,size=10)
arr2 = np.arange(0,10)
print(arr1,arr2)
plt.title('Scatter plot')
plt.xlabel('Age')
plt.ylabel('Weight')

plt.scatter(arr1,arr2)
```

```python
plt.scatter(arr1,arr2,color=['red','green','blue','yellow','skyblue','red','green','blue','yellow','skyblue'],s=60)
```

```python
plt.scatter(arr1,arr2,color=['red','green','blue','yellow','skyblue','red','green','blue','yellow','skyblue'],s=100,marker='*',edgecolors='green')
```

```python
plt.scatter(arr1,arr2,color=['red','green','blue','yellow','skyblue','red','green','blue','yellow','skyblue'],s=100,marker='*',edgecolors='green',cmap='viridis')
```
