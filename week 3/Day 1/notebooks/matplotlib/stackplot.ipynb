### STackplot

What is a Stack Plot?

    A stack plot is used to show part-to-whole relationships over time. It is like an area chart where each data series is stacked on top of the previous one.

When to Use a Stack Plot

    Best When:
        You want to visualize composition over time.
        You have multiple categories and want to see how they change.
        You want to show cumulative trends.

    Avoid When:
        You have too many categories → becomes unreadable.
        You want to compare individual values clearly → use a line plot or grouped bar chart.



```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 7, 8]
eating = [2, 2, 3, 2, 2]
working = [8, 9, 9, 10, 9]
playing = [7, 5, 6, 5, 5]

plt.stackplot(days, sleeping, eating, working, playing,
              labels=['Sleeping', 'Eating', 'Working', 'Playing'])

plt.legend(loc='upper left')
plt.title("Daily Activities Over 5 Days")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.show()
```
