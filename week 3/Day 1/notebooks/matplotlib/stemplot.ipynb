### Stem Plot

What is a stem plot?

    A stem plot is a type of plot that displays data points as lines("stems") extending from a baseline 
    (usually zero) to a marker representing the data value.

When to use stem plots?

    . To visualize discrete data or seq clearly
    . To emphasize the individual values and their positions
    . Useful in signal processing, digital signals or showing impulses


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
y = np.random.randint(1, 10, size=10)
print(x,y)

plt.stem(x, y)
plt.title("Basic Stem Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
