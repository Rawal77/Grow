### Subplot

What is a subplot?

    A subplot is a small plot inside a larger figure. It allows you to compare mutliple
    graphs side by side

Why use Subplots?

    Use when 
    
    You want to compare multiple side by side
    You want to organize multiple visualizations in one window

```python
import matplotlib.pyplot as plt 
```

```python
plt.subplot(1,2,1) # 1 row, 2 columns, 1st plot

plt.plot([1,2,3],[4,5,6])
plt.title("Left plot")

plt.subplot(1,2,2) # 1 row, 2 columns, 2nd plot
plt.plot([1, 2, 3], [6, 5, 4])
plt.title("Right Plot")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2)  # 1 row, 2 columns

# axs is a 1D NumPy array with 2 axes: [ax1, ax2]

axs[0].plot([1, 2, 3], [1, 4, 9])
axs[0].set_title("Left Plot")

axs[1].plot([1, 2, 3], [9, 4, 1])
axs[1].set_title("Right Plot")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)

# axs is now a 2D NumPy array:
# [[ax00, ax01],
#  [ax10, ax11]]

axs[0, 0].plot([1, 2], [2, 3])
axs[0, 0].set_title("Top-Left")

axs[0, 1].plot([1, 2], [3, 1])
axs[0, 1].set_title("Top-Right")

axs[1, 0].plot([1, 2], [5, 6])
axs[1, 0].set_title("Bottom-Left")

axs[1, 1].plot([1, 2], [7, 5])
axs[1, 1].set_title("Bottom-Right")

plt.tight_layout()
plt.show()
```
