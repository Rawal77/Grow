```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sample data
data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Sales": [100, 120, 90, 150],
    "Profit": [20, 25, 15, 30]
})
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(6, 4))
sns.lineplot(data=data, x="Month", y="Sales", marker="o", label="Sales")
sns.lineplot(data=data, x="Month", y="Profit", marker="s", label="Profit")
plt.title("Monthly Sales and Profit (Line Chart)")
plt.ylabel("Amount")
plt.legend()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=data, x="Month", y="Sales", palette="pastel")
plt.title("Monthly Sales (Bar Chart)")
plt.ylabel("Sales")
```

```python
plt.figure(figsize=(5, 5))
plt.pie(data["Sales"], labels=data["Month"], autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Sales Distribution by Month (Pie Chart)")
plt.show()
```
