### Heatmap

What is heatmap?

    A heatmap is a graphical representation of data where values are shown as colors in a 2D grid.
    It helps you visually analyze patterns, relationships, and concentration of values across two dimensions.

When to use a heatmap?

    You want to visualize correlation between features
    You want to analyze a 2D matrix/table
    You want to compare values accross rows and columns

When not to use it?

    Data is not structured in a 2D format (like Lists)
    You have categorical data without meaningful numeric values
    color may mislead interpretation (especially with bad colormaps)

```python
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')

# Correlation matrix
corr = df.corr(numeric_only=True)

sns.heatmap(corr,annot=True,cmap="coolwarm",fmt=".2f",linewidths=1)
```

```python
sns.heatmap(corr, annot=True, cmap="magma", cbar=True, linewidths=1, linecolor="white")
```
