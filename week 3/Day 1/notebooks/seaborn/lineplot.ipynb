### Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.DataFrame({
    "Month" : ["Jan","Feb","Mar","Apr"],
    "Sales" : [100,120,90,150]
})
```

```python
sns.set_theme(style="whitegrid")
sns.lineplot(data=df,x="Month",y="Sales",marker="o")
plt.title("Sales over months")
```

```python
df2 = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr","May","June","July","Aug"],
    "Sales": [100, 120, 90, 150, 80, 110, 95, 140],
    "Region": ["East","West","North","South","East","South","West","North"]
})
```

```python
sns.lineplot(data=df2,x="Month",y="Sales",hue="Region",marker="o") 
```
