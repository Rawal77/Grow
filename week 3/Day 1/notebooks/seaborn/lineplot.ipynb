### Line Plot

```python
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,5,1,6,7]

plt.plot(var,var_1)
```

```python
sns.lineplot({"var" : var,"var 1":var_1})
```
