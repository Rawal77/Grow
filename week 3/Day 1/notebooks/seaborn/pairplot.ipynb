### Pairplot

What is pairplot?

    A pairplot is a grid of plots showing pairwise relationships between multiple variables in a dataset.

It helps you:

    Visualize correlation between features
    Dettect patterns and clusters
    Spot Outliers
    Compare distribution across categories

Best Use case for pairplot

    You want to explore relatonships between numeric variables
    You want to compare classes using hue (color coding)
    You want a quick visual summary of dataset

When to Avoid Pairplot

    Avoid it when:

     Dataset has too many columns (e.g. >10) → gets messy and slow

     You want only one or two variable relationships → use scatter or regplot

     Categorical-only data → pairplot is for numerical data
