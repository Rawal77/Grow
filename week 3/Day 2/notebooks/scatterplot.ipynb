### Scatter plot

```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.set_theme(style="whitegrid")
sns.scatterplot(data=titanic,x='age',y='fare',hue='pclass',palette='Set2')
plt.suptitle("Age vs Fare Colored by passenger Class",y=1.02)
plt.xlabel("Age")
plt.ylabel("Fare")
```

```python
sns.boxplot(data=titanic, x='pclass', y='age', palette='pastel')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
```

```python
sns.boxplot(data=titanic, x='sex', y='age', palette='coolwarm')
plt.title('Age Distribution by Gender')
plt.xlabel('Sex')
plt.ylabel('Age')
```
