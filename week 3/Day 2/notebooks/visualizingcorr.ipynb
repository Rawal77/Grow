### Visualizing Correlation

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
penguins = sns.load_dataset('penguins')
```

```python
tips
```

```python
tips_corr = tips.corr(numeric_only=True)
print("Correlation for the Tips Dataset: ")
tips_corr
```

```python
sns.heatmap(tips_corr,annot=True,fmt='.2f',cmap='coolwarm')
```

```python
penguins
```

```python
penguins_corr = penguins.corr(numeric_only=True)
penguins_corr
```

```python
sns.heatmap(penguins_corr,annot=True,fmt=".2f",cmap="coolwarm")
```
