### Bar line Plot

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.barplot(x='pclass',y="fare",data=titanic,hue="sex", estimator="mean")
plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
```

```python
sns.barplot(x='embark_town', y='survived', data=titanic)
plt.title('Survival Rate by Embark Town')
plt.ylabel('Survival Rate')
plt.show()
```

```python
age_survival = titanic.groupby('age')['survived'].mean().reset_index()

sns.lineplot(x="age", y="survived",data=age_survival,markers="o",color="green")
```
