### Working on probability

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
# 1. Simulate a Coin Toss 100 times

coin_tosses = np.random.choice(['Heads', 'Tails'], size=100)
num_heads = np.sum(coin_tosses == 'Heads')
prob_heads = num_heads / 100
```

```python
# 2. Simulate Rolling Two Dice 1000 times and estimate P(sum=7)

die1 = np.random.randint(1, 7, size=1000)
die2 = np.random.randint(1, 7, size=1000)
sum_dice = die1 + die2
prob_sum_7 = np.sum(sum_dice == 7) / 1000
```

```python
# 3. Create Probability Table for a Single Die Roll

outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6
prob_table = pd.DataFrame({
    'Outcome': outcomes,
    'Probability': probabilities
})
```

```python
# 4. Visualize the Probability Table

plt.figure(figsize=(7, 4))
sns.barplot(x='Outcome', y='Probability', data=prob_table, palette='pastel')
plt.title('Probability Distribution of a Fair Die Roll')
plt.ylim(0, 0.2)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
