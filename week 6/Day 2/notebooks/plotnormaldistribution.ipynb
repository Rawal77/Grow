### Plotting Normal Distribution

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm
```

```python
# Parameters for the Normal Distribution

mu = 170        # mean
sigma = 10      # standard deviation
```

```python
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  
y = norm.pdf(x, loc=mu, scale=sigma)              
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='darkblue', linewidth=2)
plt.title('Normal Distribution ( = 170,  = 10)', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)

# Mean line
plt.axvline(mu, color='red', linestyle='--', label='Mean ()')
plt.legend()
plt.grid(True)
plt.show()
```

```python
prob_less_than_180 = norm.cdf(180, loc=mu, scale=sigma)
print(f"P(Height < 180 cm): {prob_less_than_180:.4f}")
```

```python
prob_160_to_180 = norm.cdf(180, mu, sigma) - norm.cdf(160, mu, sigma)
print(f"P(160 < Height < 180): {prob_160_to_180:.4f}")
```
