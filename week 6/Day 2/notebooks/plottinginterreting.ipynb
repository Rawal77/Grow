```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm
```

```python
# Step 1: Simulate Coin Tosses
n_tosses = 10000
coin_tosses = np.random.choice([0, 1], size=n_tosses)  # 0 = Tails, 1 = Heads
heads_prob = np.mean(coin_tosses)
```

```python

# Step 2: Simulate Dice Rolls
n_rolls = 10000
dice_rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=n_rolls)
six_prob = np.sum(dice_rolls == 6) / n_rolls
```

```python
# Step 3: Generate Binomial Distribution
n, p = 20, 0.5  # 20 trials, 0.5 probability
x_binom = np.arange(0, n+1)
y_binom = binom.pmf(x_binom, n, p)
```

```python
# Step 4: Generate Normal Distribution
mu, sigma = 170, 10
x_norm = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_norm = norm.pdf(x_norm, mu, sigma)
```

```python
# Step 5: Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Binomial Distribution Plot
ax1.bar(x_binom, y_binom, color='skyblue', edgecolor='black')
ax1.set_title('Binomial Distribution (n=20, p=0.5)', fontsize=14)
ax1.set_xlabel('Number of Successes', fontsize=12)
ax1.set_ylabel('Probability', fontsize=12)
ax1.grid(True)

# Normal Distribution Plot with shaded   1
ax2.plot(x_norm, y_norm, color='darkblue', linewidth=2)
ax2.fill_between(x_norm, y_norm, where=(x_norm >= mu - sigma) & (x_norm <= mu + sigma),
                 color='skyblue', alpha=0.5, label='  1')
ax2.axvline(mu, color='red', linestyle='--', label='Mean ()')
ax2.set_title('Normal Distribution (=170, =10)', fontsize=14)
ax2.set_xlabel('Value', fontsize=12)
ax2.set_ylabel('Density', fontsize=12)
ax2.legend()
ax2.grid(True)

plt.tight_layout()

(heads_prob, six_prob)
```

### probability of getting 0 to 20 heads in 20 tosses.

    Peak at 10 heads (the mean).

    Symmetric shape due to p = 0.5.

    Variance = np(1-p) = 5, Standard deviation  2.24



###  Normal Distribution (=170, =10):

    Bell-shaped curve centered at  = 170.

    Shaded region represents   1 (160 to 180), containing ~68% of values.

    Tails represent rare/extreme outcomes.

    Spread () controls the width: higher   flatter and wider curve.
