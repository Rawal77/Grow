### Performing Confidence Interval

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
np.random.seed(100)
sample_size = 50
population_mean = 5  # Just a hypothetical true mean (can be ignored here)
population_std = 2
```

```python
# Generate a sample
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Calculate sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # sample standard deviation
```

```python
confidence_level = 0.95
alpha = 1 - confidence_level
df = sample_size - 1

# t critical value for 95% confidence and df degrees of freedom
t_crit = stats.t.ppf(1 - alpha/2, df)

# Standard Error (SE)
SE = sample_std / np.sqrt(sample_size)

# Margin of Error (ME)
ME = t_crit * SE

# Confidence Interval
CI_lower = sample_mean - ME
CI_upper = sample_mean + ME

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({CI_lower:.2f}, {CI_upper:.2f})")
```

```python
plt.figure(figsize=(6,6))

# Plot sample mean as a point
plt.errorbar(x=1, y=sample_mean, yerr=ME, fmt='o', color='blue', 
             ecolor='red', elinewidth=3, capsize=10, label='95% Confidence Interval')

plt.xlim(0.5, 1.5)
plt.xticks([1], ['Sample Mean'])
plt.ylabel('Value')
plt.title('Sample Mean with 95% Confidence Interval')

plt.legend()
plt.grid(True)
plt.show()
```

The range where the true population mean is likely to lie with 95% confidence.
