### Sampling Distribution by Simulating 1,000 Sample Means

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# Simulate population: 100,000 individuals, average height ~170 cm, std dev ~10 cm
population = np.random.normal(loc=170, scale=10, size=100000)

# Calculate true population mean
true_mean = np.mean(population)
print(f"True population mean: {true_mean:.2f} cm")
```

```python
# Draw 1,000 samples of size n=50 and compute sample means



sample_size = 50
num_samples = 1000
sample_means = []

for i in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(sample_means, bins=30, kde=True, color='skyblue')

plt.axvline(true_mean, color='red', linestyle='dashed', linewidth=2, label=f"True Population Mean ({true_mean:.2f})")

plt.title("Sampling Distribution of the Sample Mean (n=50, 1000 samples)")
plt.xlabel("Sample Mean Height (cm)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

I am 95% confident that the true average height of all individuals in the population lies between 165 cm and 171 cm.

    This means if repeating this sampling process many times, about 95% of the intervals  calculate will contain the true population mean.
    

```python
import numpy as np
import pandas as pd

# Seed for reproducibility
np.random.seed(42)

# Simulate population: height in cm (mean = 170, std dev = 10)
population = pd.DataFrame({
    'Height': np.random.normal(loc=170, scale=10, size=10000)
})

# Calculate population mean
population_mean = population['Height'].mean()
print(f"Population Mean Height: {population_mean:.2f} cm")
```

```python
# Choose sampling interval k
k = 10

# Random starting index between 0 and k-1
start = np.random.randint(0, k)

# Systematic sampling: select every k-th row starting from 'start'
systematic_sample = population.iloc[start::k]

# Calculate sample mean
sample_mean = systematic_sample['Height'].mean()
print(f"Systematic Sample Mean Height: {sample_mean:.2f} cm")
```

```python
difference = abs(population_mean - sample_mean)
print(f" Difference between Population and Sample Mean: {difference:.4f} cm")
```
