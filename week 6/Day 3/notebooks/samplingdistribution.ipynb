### Performing sampling distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For reproducibility
np.random.seed(42)
```

```python
# Create a non-normal (skewed) population data example:
# Using exponential distribution to show CLT effect clearly
population_data = np.random.exponential(scale=2, size=10000)


data = pd.DataFrame({'Value': population_data})


# Check basic info
print(f"Population Mean: {data['Value'].mean():.2f}")
print(f"Population Std Dev: {data['Value'].std():.2f}")
```

```python
sample_size = 50
num_samples = 30
sample_means = []

for i in range(num_samples):
    sample = data['Value'].sample(n=sample_size, replace=False, random_state=42 + i)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

sample_means = np.array(sample_means)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(sample_means, kde=True, color='skyblue', bins=10)

plt.title("Sampling Distribution of Sample Means (n=50, 30 samples)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.axvline(sample_means.mean(), color='red', linestyle='dashed', linewidth=2, label=f'Mean of sample means: {sample_means.mean():.2f}')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data['Value'], bins=30, color='orange', kde=True)

plt.title("Population Distribution (Exponential - Skewed)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
