```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Step 1: Create a skewed population (Exponential Distribution)
population = np.random.exponential(scale=10, size=100000)

# Population statistics
pop_mean = np.mean(population)
pop_std = np.std(population)

print(f"Population Mean: {pop_mean:.2f}")
print(f"Population Std Dev: {pop_std:.2f}")

# Step 2: Sampling function
def sampling_distribution(sample_size, num_samples=1000):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))
    return sample_means

# Step 3: Plotting sampling distributions for different sample sizes
sample_sizes = [5, 30, 100]
plt.figure(figsize=(15, 4))

for i, n in enumerate(sample_sizes, 1):
    sample_means = sampling_distribution(sample_size=n)
    plt.subplot(1, 3, i)
    sns.histplot(sample_means, kde=True, bins=30, color='skyblue')
    plt.axvline(pop_mean, color='red', linestyle='--', label='Population Mean')
    plt.title(f'Sample Size = {n}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.legend()

plt.suptitle("Central Limit Theorem Demonstration", fontsize=16)
plt.tight_layout()
plt.show()
```
